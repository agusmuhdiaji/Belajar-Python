# Dealing with Missing Value (NAs)

```python
import numpy as np
import pandas as pd
```

```python
#create sampel data
df=pd.DataFrame([[1,np.nan,2],[2,3,5],[np.nan,4,6]]) #terdapat value yang hilang atau tidak ada dalam sel
#kita perlu menghilangkan NA atau value yang hilang
```

```python
print(df)
```

```python
#mengecek jika terdapat NA dalam data frame
print(df.isnull())
#True berarti terdapat NA dalam data frame
```

```python
#drop atau menghilangkan nilai NA
#semua baris dengan sel yang terdapat NA akan hilang
print(df.dropna())
```

```python
#drop atau menghilangkan kolom yang terdapat nilai NA
print(df.dropna(axis=1))
```

```python
#thresh parameter untuk menspesifikan minimum number dari nilai non-null untuk baris dan kolom
print(df.dropna(thresh=2))
```

## mengisi nilai NA

```python
#mengisi nilai NA dengan 0 atau zero
print(df.fillna(0))
```

```python
#mengisi nilai NA dengan mengambil nilai sebelumnya untuk mengisi nilai setelahnya atau forwardfill
print(df.fillna(method="ffill"))
```

```python
#mengisi nilai NA berdasarkan kolom
print(df.fillna(method="ffill",axis=1))
```

```python
#mengisi nilai NA dengan mengambil nilai setelahnya untuk mengisi nilai sebelumnya atau backfill
print(df.fillna(method="bfill"))
```

## database type merging

```python
import pandas as pd
```

```python
gfp=pd.read_csv("D:\Data Science\scriptsLecture\section5\GlobalFirePower.csv")
gfp.head(5)
```

```python
gdp=pd.read_csv("D:\Data Science\scriptsLecture\section5\countryGDP.csv")
gdp.head(5)
```

```python
#Keep every row in the left dataframe. Where there are missing values of the “on” variable in the right dataframe, add empty
y = pd.merge(gdp,gfp[['Country', 'Manpower Available', 'Railway Coverage (km)']],
                 on='Country',how='left')
y.head(6)
```

```python
gfp['Country'].isin(gdp['Country']).value_counts() #126 values of gdp appear in gfp too
```

```python
y = pd.merge(gdp,gfp[['Country', 'Manpower Available', 'Railway Coverage (km)']],
                 on='Country',how='right')
y.tail(7)

# Keep every row in the right dataframe. Where there are missing values of the “on” variable in the left column, add empty
```

```python
y = pd.merge(gdp,gfp[['Country', 'Manpower Available', 'Railway Coverage (km)']],
                 on='Country',how='outer',indicator=True)
y.head(7)
#A full outer join returns all the rows from the left dataframe, 
#all the rows from the right dataframe, and matches up rows where possible, with NaNs elsewhere
#indicator will tell us how the joining was done
```
